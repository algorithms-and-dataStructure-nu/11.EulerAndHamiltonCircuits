# 11 Euler閉路とHamilton閉路: Eulerian Circuit and Hamiltonian Circuit

```python
from collections.abc import Hashable
from typing import Hashable
import networkx as nx
```

## Euler閉路を列挙: Enumerating Eulerian Circuits
- Euler閉路は、グラフのすべての辺をちょうど一度ずつ通る閉路
- 全ての頂点の次数(接続する辺の数)が偶数である必要がある
- Euler閉路を探すには、再帰的に辺をたどる方法がある
- 閉路が見つからず戻る場合には、探索履歴をリセットして別の経路を試す
- 閉路が見つかったら、それを保存して次の探索に進む
----
- An Eulerian circuit is a closed path that visits every edge of a graph exactly once
- Every vertex must have an even degree (number of edges connected to it) for an Eulerian circuit to exist
- To find Eulerian circuits, a recursive approach can be used to traverse edges
- If a path does not lead to a solution, the search history is reset and another path is tried
- When a circuit is found, it is saved and the search continues for the next circuit

```python
def enumerate_Euler_main(G:nx.Graph, s:Hashable)->list[tuple[Hashable, Hashable]]:
    """
    Euler閉路を列挙する関数: Function to enumerate Eulerian circuits:

    Parameters:
    G: nx.Graph - 列挙するグラフ: The graph to enumerate
    s: Hashable - Euler閉路の開始点: Starting point for the Eulerian circuit
    """
    G_copy = G.copy()  # グラフのコピーを作成: Create a copy of the graph
    euler_paths = []
    
    def find_euler_path(v:Hashable, path:list[tuple[Hashable, Hashable]], ss:Hashable):
        # 終了条件: 全ての辺を訪れ、開始点に戻る: Base case: All edges visited and back to start
        if len(path) == G_copy.number_of_edges() and v == ss:
            euler_paths.append(path.copy())
            return
        
        for neighbor in list(G_copy.neighbors(v)):# 隣接する頂点を探索: Explore adjacent vertices
            if (v, neighbor) not in path and (neighbor, v) not in path:
                # まだ訪れていない辺を選択: Choose an edge that has not been visited
                path.append((v, neighbor))
                find_euler_path(neighbor, path, ss)
                path.pop()
    
    find_euler_path(s, [], s)
    return euler_paths

def euler_example():
    """ 
    Eulerを列挙するグラフの例: Example graph for enumerating Eulerian circuits
    """
    G = nx.Graph()
    G.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4), (2, 5), (3, 4), (3, 5)])
    pos = {1: (0, 0), 2: (-1, -1), 3: (1, -1), 4: (-1, -2), 5: (1, -2)}
    nx.set_node_attributes(G, pos, "pos")
    return G

g = euler_example()
pos = nx.get_node_attributes(g, "pos")
nx.draw(g, with_labels=True, node_color="lightblue", edge_color="black", pos=pos)
paths = enumerate_Euler_main(g, 1)
print("Eulerian Circuits:", len(paths))
for path in paths:
    print(path)
```

## Hamilton閉路を列挙: Enumerating Hamiltonian Circuits
- Hamilton閉路は、グラフのすべての頂点をちょうど一度ずつ通る閉路
- Hamilton閉路を探すには、再帰的に頂点をたどる方法がある
- 閉路が見つからず戻る場合には、探索履歴をリセットして別の経路を試す
- 閉路が見つかったら、それを保存して次の探索に進む

----
- A Hamiltonian circuit is a closed path that visits every vertex of a graph exactly once
- To find Hamiltonian circuits, a recursive approach can be used to traverse vertices
- If a path does not lead to a solution, the search history is reset and another path is tried
- When a circuit is found, it is saved and the search continues for the next circuit

```python
def enumerate_Hamiltonian_main(G:nx.Graph, s:Hashable)->list[list[Hashable]]:
    """
    Hamilton閉路を列挙する関数: Function to enumerate Hamiltonian circuits:

    Parameters:
    G: nx.Graph - 列挙するグラフ: The graph to enumerate
    s: Hashable - Hamilton閉路の開始点: Starting point for the Hamiltonian circuit
    """
    G_copy = G.copy()  # グラフのコピーを作成: Create a copy of the graph
    hamilton_paths = []
    
    def find_hamilton_path(v:Hashable, path:list[Hashable], ss:Hashable):
        # 終了条件: 全ての頂点を訪れ、開始点に戻る: Base case: All vertices visited and back to start
        if len(path) == G_copy.number_of_nodes() and v == ss:
            hamilton_paths.append(path.copy())
            return
        
        for neighbor in G_copy.neighbors(v):# 隣接する頂点を探索: Explore adjacent vertices
            if neighbor not in path:
                # まだ訪れていない頂点を選択: Choose a vertex that has not been visited
                path.append(neighbor)
                find_hamilton_path(neighbor, path, ss)
                path.pop()
    
    find_hamilton_path(s, [s], s)
    return hamilton_paths

def hamilton_example():
    """ 
    Hamiltonを列挙するグラフの例: Example graph for enumerating Hamiltonian circuits
    """
    G = nx.Graph()
    G.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4), (2, 5), (3, 4), (3, 5)])
    pos = {1: (0, 0), 2: (-1, -1), 3: (1, -1), 4: (-1, -2), 5: (1, -2)}
    nx.set_node_attributes(G, pos, "pos")
    return G

g = hamilton_example()
pos = nx.get_node_attributes(g, "pos")  
nx.draw(g, with_labels=True, node_color="lightblue", edge_color="black", pos=pos)
paths = enumerate_Hamiltonian_main(g, 1)
print("Hamiltonian Circuits:", len(paths))
for path in paths:
    print(path)
```
