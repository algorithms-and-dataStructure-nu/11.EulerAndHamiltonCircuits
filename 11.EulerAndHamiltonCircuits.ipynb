# 11 Euler閉路とHamilton閉路: Eulerian Circuit and Hamiltonian Circuit

```python
from collections.abc import Hashable
import networkx as nx
import random
```

## Euler閉路を列挙: Enumerating Eulerian Circuits
- Euler閉路は、無向グラフのすべての辺をちょうど一度ずつ通る閉路
- 全ての頂点の次数(接続する辺の数)が偶数である必要がある
- Euler閉路を探すには、再帰的に辺をたどる方法がある
- 閉路が見つからず戻る場合には、探索履歴をリセットして別の経路を試す
- 閉路が見つかったら、それを保存して次の探索に進む
----
- An Eulerian circuit is a closed path that visits every edge of a undirected graph exactly once
- Every vertex must have an even degree (number of edges connected to it) for an Eulerian circuit to exist
- To find Eulerian circuits, a recursive approach can be used to traverse edges
- If a path does not lead to a solution, the search history is reset and another path is tried
- When a circuit is found, it is saved and the search continues for the next circuit

```python
def enumerate_Euler_main(G:nx.Graph, s:Hashable)->list[list[tuple[Hashable, Hashable]]]:
    """
    Euler閉路を列挙する関数: Function to enumerate Eulerian circuits:

    Parameters:
    G: nx.Graph - 列挙する無向グラフ: The undirected graph to enumerate
    s: Hashable - Euler閉路の開始点: Starting point for the Eulerian circuit
    """
    G_copy:nx.Graph = G.copy()  # グラフのコピーを作成: Create a copy of the graph
    euler_paths:list[list[tuple[Hashable, Hashable]]] = []
    
    def find_euler_path(v:Hashable, path:list[tuple[Hashable, Hashable]], ss:Hashable):
        """
        再帰的に一つのEuler閉路を探索する関数: Recursive function to explore one Eulerian circuit

        Parameters:
        v: Hashable - 現在の頂点: Current vertex
        path: list[tuple[Hashable, Hashable]] - 現在の経路: Current path
        ss: Hashable - 開始点: Starting point
        """
        # 終了条件: 全ての辺を訪れ、開始点に戻る: Base case: All edges visited and back to start
        if len(path) == G_copy.number_of_edges() and v == ss:
            euler_paths.append(path.copy())
            return
        
        for neighbor in list(G_copy.neighbors(v)):# 隣接する頂点を探索: Explore adjacent vertices
            if (v, neighbor) not in path and (neighbor, v) not in path:
                # まだ訪れていない辺を選択: Choose an edge that has not been visited
                path.append((v, neighbor))
                find_euler_path(neighbor, path, ss)
                path.pop()# バックトラック時に辺を削除: Remove the edge during backtracking
    
    # Euler閉路の探索を開始: Start exploring for Eulerian circuits
    find_euler_path(s, [], s)
    return euler_paths

def euler_example()->nx.Graph:
    """ 
    Eulerを列挙するグラフの例: Example graph for enumerating Eulerian circuits
    """
    G = nx.Graph()
    G.add_edges_from([('A', 'B'), ('A', 'C'), 
                      ('B', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'D'), ('C', 'E')])
    pos = {'A': (0, 0), 'B': (-1, -1), 
           'C': (1, -1), 'D': (-1, -2), 'E': (1, -2)}
    nx.set_node_attributes(G, pos, "pos")
    return G

def euler_example2()->nx.Graph:
    """ 
    Eulerを列挙するグラフの例: Example graph for enumerating Eulerian circuits
    """
    G = nx.Graph()
    G.add_edges_from([('A', 'B'), ('A', 'C'), ('A', 'D'), ('A', 'E'),
                      ('B', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'D'), ('C', 'E'), ('D', 'E')])
    pos = {'A': (0, 0), 'B': (-1, -1), 
           'C': (1, -1), 'D': (-1, -2), 'E': (1, -2)}
    nx.set_node_attributes(G, pos, "pos")
    return G

# Graphを作成: Create the graph
g:nx.Graph = euler_example()
pos = nx.get_node_attributes(g, "pos")
nx.draw(g, with_labels=True, node_color="lightblue", edge_color="black", pos=pos)
# Euler閉路を列挙: Enumerate Eulerian circuits
paths: list[list[tuple[Hashable, Hashable]]] = enumerate_Euler_main(g, 'A')
print(f'number of Eulerian Circuits: {len(paths)}')
print("Eulerian Circuits:", len(paths))
for path in paths:
    print(path)
```

## Hamilton閉路を列挙: Enumerating Hamiltonian Circuits
- Hamilton閉路は、無向グラフのすべての頂点をちょうど一度ずつ通る閉路
- Hamilton閉路を探すには、再帰的に頂点をたどる方法がある
- 閉路が見つからず戻る場合には、探索履歴をリセットして別の経路を試す
- 閉路が見つかったら、それを保存して次の探索に進む

----
- A Hamiltonian circuit is a closed path that visits every vertex of an undirected graph exactly once
- To find Hamiltonian circuits, a recursive approach can be used to traverse vertices
- If a path does not lead to a solution, the search history is reset and another path is tried
- When a circuit is found, it is saved and the search continues for the next circuit

```python
def enumerate_Hamiltonian_main(G:nx.Graph, s:Hashable)->list[list[Hashable]]:
    """
    Hamilton閉路を列挙する関数: Function to enumerate Hamiltonian circuits:

    Parameters:
    G: nx.Graph - 列挙するグラフ: The graph to enumerate
    s: Hashable - Hamilton閉路の開始点: Starting point for the Hamiltonian circuit
    """
    G_copy: nx.Graph = G.copy()  # グラフのコピーを作成: Create a copy of the graph
    hamilton_paths: list[list[Hashable]] = []
    
    def find_hamilton_path(v:Hashable, path:list[Hashable], ss:Hashable):
        """
        再帰的に一つのHamilton閉路を探索する関数: Recursive function to explore one Hamiltonian circuit

        Parameters:
        v: Hashable - 現在の頂点: Current vertex
        path: list[Hashable] - 現在の経路: Current path
        ss: Hashable - 開始点: Starting point
        """
        # 終了条件: 全ての頂点を訪れ、開始点に戻る: Base case: All vertices visited and back to start
        if len(path) == G_copy.number_of_nodes():
            if ss in G_copy.neighbors(v):  
                # 最後の頂点から開始点に戻れるか: Check if we can return to the start from the last vertex
                hamilton_paths.append(path.copy())
            return
        
        for neighbor in G_copy.neighbors(v):# 隣接する頂点を探索: Explore adjacent vertices
            if neighbor not in path:
                # まだ訪れていない頂点を選択: Choose a vertex that has not been visited
                path.append(neighbor)
                find_hamilton_path(neighbor, path, ss)
                path.pop()
    # Hamilton閉路の探索を開始: Start exploring for Hamiltonian circuits
    find_hamilton_path(s, [s], s)
    return hamilton_paths

def hamilton_example()->nx.Graph:
    """ 
    Hamiltonを列挙するグラフの例: Example graph for enumerating Hamiltonian circuits
    """
    G: nx.Graph = nx.Graph()
    G.add_edges_from([('A', 'B'), ('A', 'C'),
                    ('B', 'C'), ('B', 'D'), ('B', 'E'),
                    ('C', 'D'), ('C', 'E'), ('D', 'E')])
    pos = {'A': (0, 0), 'B': (-1, -1), 'C': (1, -1), 'D': (-1, -2), 'E': (1, -2)}
    nx.set_node_attributes(G, pos, "pos")
    return G

def hamilton_example2()->nx.Graph:
    """ 
    Hamiltonを列挙するグラフの例: Example graph for enumerating Hamiltonian circuits
    """
    G: nx.Graph = nx.Graph()
    G.add_edges_from([('A', 'B'), ('A', 'C'),
                    ('B', 'C'), ('B', 'D'), ('B', 'E'), ('B', 'G'),
                    ('C', 'D'), ('C', 'E'), ('C', 'F'), ('D', 'E'), ('D', 'F'),
                    ('D', 'G'), ('E', 'F'), ('E', 'G'),('F', 'G')])
    pos = {'A': (0, 0), 'B': (-1, -1), 'C': (1, -1), 'D': (-1, -2), 'E': (1, -2), 'F': (-.5, -3), 'G': (.5, -3)}
    nx.set_node_attributes(G, pos, "pos")
    return G

# Graphを作成: Create the graph
g = hamilton_example2()
pos = nx.get_node_attributes(g, "pos")  
nx.draw(g, with_labels=True, node_color="lightblue", edge_color="black", pos=pos)
# Hamilton閉路を列挙: Enumerate Hamiltonian circuits
paths = enumerate_Hamiltonian_main(g, 'A')

print("Hamiltonian Circuits:", len(paths))
for path in paths:
    path.append(path[0])  # 最後の頂点から開始点に戻る: Return to the start from the last vertex
    print(path)

selected_path = random.choice(paths) if paths else None 
if selected_path:
    selected_path.append(selected_path[0])  # 最後の頂点から開始点に戻る: Return to the start from the last vertex
    nx.draw_networkx_edges(g, edge_color='red', width=2, pos=pos, edgelist=[(selected_path[i], selected_path[i+1]) for i in range(len(selected_path)-1)])
```
